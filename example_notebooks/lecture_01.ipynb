# how to lie with means

```python
their_results = 120

winning_number = their_results + 5

np.random.seed(90)
my_results_1 = np.random.randint(30,80,4)
my_results_1 = np.append(my_results_1,winning_number*(len(my_results_1)+1)-sum(my_results_1))
my_results_1

my_results_2 = np.append(np.random.randint(0,30,6), np.random.randint(100,140,6))
my_results_2 = (winning_number - np.mean(my_results_2)) + my_results_2

print(their_results, str(np.mean(my_results_1)), str(np.mean(my_results_2)),)
```

```python
plt.xkcd(True)

plt.figure(dpi=150)
plt.scatter(1,np.mean(their_results), sizes=[500], marker="*")
plt.scatter(2, np.mean(my_results_1), sizes=[500], marker="*")
plt.scatter(3, np.mean(my_results_2), sizes=[500], marker="*")
plt.xlim([0.5,3.5])
plt.xticks([1,2,3],labels=['theirs','ours','also ours'])
plt.ylabel("awesome")
plt.xlabel("method")
plt.title("Our Methods are Better than Theirs (Means)")
plt.tight_layout()
plt.savefig("../images/lecture01/data_means.png", bbox_inches="tight")
plt.show()

plt.figure(dpi=150)
plt.errorbar(1, np.mean(their_results), np.std(their_results), capsize=0.2, marker="*")
plt.errorbar(2, np.mean(my_results_1), np.std(my_results_1), capsize=0.2, marker="o")
plt.errorbar(3, np.mean(my_results_2), np.std(my_results_2), capsize=0.2, marker="o")
plt.xlim([0.5,3.5])
plt.xticks([1,2,3],labels=['theirs','ours','also ours'])
plt.ylabel("awesome")
plt.xlabel("method")
plt.title("Our Methods are Better than Theirs(?) \n(With Standard Deviations)")
plt.tight_layout()
plt.savefig("../images/lecture01/data_errorbars.png", bbox_inches="tight")
plt.show()


plt.figure(dpi=150)
plt.scatter(1,their_results, marker="*")
plt.scatter(2*np.ones(len(my_results_1)), my_results_1)
plt.scatter(3*np.ones(len(my_results_2)), my_results_2)
plt.xlim([0.5,3.5])
plt.xticks([1,2,3],labels=['theirs','ours','also ours'])
plt.ylabel("awesome")
plt.xlabel("method")
plt.title("Our Methods are Better than Theirs(??) (Raw)")
plt.tight_layout()
plt.savefig("../images/lecture01/data_raw.png", bbox_inches="tight")
plt.show()
```

```python
our_lab_1 = np.random.uniform(0,10,11)
lab_median = np.median(our_lab_1)

our_lab_2 = np.random.uniform(lab_median-1,lab_median,5)
our_lab_2 = np.append(our_lab_2, lab_median)
our_lab_2 = np.append(our_lab_2, np.random.uniform(lab_median,lab_median+1,5))

our_lab_3 = np.random.uniform(0,0.5,5)
our_lab_3 = np.append(our_lab_3, lab_median)
our_lab_3 = np.append(our_lab_3, np.linspace(lab_median,10,5))

plt.figure(dpi=200)
[plt.scatter([], []) for i in range(3)]    # hack to advance the color cycle
plt.plot([1,3],lab_median*np.ones(2),'k')
plt.scatter(np.ones(len(our_lab_1)), our_lab_1)
plt.scatter(2*np.ones(len(our_lab_2)), our_lab_2)
plt.scatter(3*np.ones(len(our_lab_3)), our_lab_3)
plt.xlim([0.5,3.5])
plt.xticks([1,2,3],labels=['Model A','Model B','Model C'])
plt.ylabel("awesome")
plt.title("Amount of Awesome Produced (Raw)")
plt.tight_layout()
plt.savefig("../images/lecture01/medians_raw.png", bbox_inches="tight")
plt.show()
```
